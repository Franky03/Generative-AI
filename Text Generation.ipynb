```python
import transformers
from transformers import pipeline
```

```python
def get_hyperparameters():
  return {
    #"max_length": 100,
    "do_sample": True,
    "temperature": 0.8,
    "top_k": 50,
    "top_p": 0.85,
    "repetition_penalty": 1.0,
    "length_penalty": 1.0,
    "num_return_sequences": 3,
    "early_stopping": False,
    "pad_token_id": 50256
  }

```

```python
get_hyperparameters()
```

```python
def initialize_generator(model_name="gpt2-medium"):
  try:
    generator = pipeline("text-generation", model=model_name)
    return generator
  except:
    print("Error on initialization!")
    return None


def generate_text(generator, prompt, params):
    try:
        result = generator(prompt, **params)
        return result
    except Exception as e:
        print(f"Error on text generation: {e}")
        return None
```

```python
def main():
  prompt = input("Prompt: ")

  parameters = get_hyperparameters()

  generator = initialize_generator("gpt2")
  results = generate_text(generator, prompt, parameters)
  return results
```

```python


results = main()
for i, result in enumerate(results):
  print(f"\nResult {i + 1}:\n{result['generated_text']}")
```
